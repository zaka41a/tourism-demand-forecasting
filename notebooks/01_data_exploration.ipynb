```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("../data/raw/tourism_data.csv", parse_dates=["date"])

df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure()
plt.plot(df["date"], df["bookings"])
plt.xlabel("Date")
plt.ylabel("Number of Bookings")
plt.title("Tourism Demand Over Time")
plt.show()
```

```python
plt.figure()
plt.scatter(df["temperature"], df["bookings"])
plt.xlabel("Temperature")
plt.ylabel("Bookings")
plt.title("Temperature vs Bookings")
plt.show()
```

## Key Insights
- Demand shows temporal variation.
- Higher temperatures tend to correlate with increased bookings.
- Dataset is clean and usable for modeling.
